# RegressionVisualizer3D

This a demonstration of the RegressionVisualizer3D package. The package is for visualizing linear regression, Poisson regression and binary logistic regression in 3D, for teaching purposes.

The PyPI project directory for the `RegressionVisualizer3D` package is here: https://pypi.org/project/RegressionVisualizer3D/

The package can be installed with `pip install RegressionVisualizer3D`.


# Model plot (default parameters)

The `model_plot()` function generates visualizations of several types of regression model. It uses a data-generating process with specific (user-controlled) parameters to generate population data. A sample (of a size controlled by the user) is then randomly drawn from the population data, and a regression model is fit to that sample. If the user does not supply population parameters, defaults are used.

The cell below runs the `model_plot()` function with its default parameters.

```python
# importing the package
import RegressionVisualizer3D

# generate and plot the data and regression surfaces, using all default settings
RegressionVisualizer3D.model_plot()
```

# Poisson regression plot (default parameters)

The cell below runs the `model_plot()` function to visualize a Poisson regression, using the default parameters.

```python
RegressionVisualizer3D.model_plot(model_type = 'poisson_regression')
```

# Binary logistic regression plot (default paramters)

The cell below runs the `model_plot()` function to visualize a Binary logistic regression, using the default parameters.

```python
RegressionVisualizer3D.model_plot(model_type = 'binary_logistic_regression')
```

# Model plots with custom paramters

The cells below show how the user can tweak the parameters of the data-generating process, to see the effect this has on the population and sample data, and on the regression surfaces.

## Linear regression

```python
# with custom parameters, controlling the data-generating process
RegressionVisualizer3D.model_plot(model_type = 'linear_regression',
                       intercept = 20,
                       predictor1_slope = 0.8, 
                       predictor2_slope = 3, 
                       interaction_slope = 0.9,
                       sample_size = 30,
                       population_size = 3000,
                       error_sd = 20,
                       predictor1_name = 'Predictor 1',
                       predictor2_name = 'Predictor 2',
                       outcome_variable_name = 'Outcome Variable',
                       axis_size = 10)
```

## Poisson regression

```python
# with custom parameters, controlling the data-generating process
RegressionVisualizer3D.model_plot(model_type = 'poisson_regression',
                       intercept = 0,
                       predictor1_slope = 0, 
                       predictor2_slope = 0.03, 
                       interaction_slope = 0.02,
                       sample_size = 30,
                       population_size = 2500,
                       error_sd = 0.1,
                       predictor1_name = 'Predictor 1',
                       predictor2_name = 'Predictor 2',
                       outcome_variable_name = 'Outcome Variable',
                       axis_size = 10)
```

## Logistic regression

```python
# with custom parameters, controlling the data-generating process
RegressionVisualizer3D.model_plot(model_type = 'binary_logistic_regression',
                       intercept = 0,
                       predictor1_slope = 0.9, 
                       predictor2_slope = 0.6, 
                       interaction_slope = 0.001,
                       sample_size = 50,
                       population_size = 1800,
                       error_sd = 5,
                       predictor1_name = 'Predictor 1',
                       predictor2_name = 'Predictor 2',
                       outcome_variable_name = 'Outcome Variable',
                       axis_size = 10)
```
